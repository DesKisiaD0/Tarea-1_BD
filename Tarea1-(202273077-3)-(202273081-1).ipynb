## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Director"

```python
%%sql
CREATE TABLE director (
    id_director INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45),
    edad INT

);
```

### Tabla "Actor"

```python
%%sql
CREATE TABLE actor ( 
    id_actor INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INTEGER

);
```

### Tabla "Pelicula"

```python
%%sql
CREATE TABLE pelicula (
    id_pelicula INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    genero VARCHAR(45) NOT NULL,
    duracion INTEGER,
    fecha_publicacion DATE,
    id_director INTEGER,
    FOREIGN KEY (id_director) REFERENCES director(id_director)
);
```

### Tabla "actor_participa_en_pelicula"

```python
%%sql
CREATE TABLE actor_participa_en_pelicula (
    id_actor INTEGER,
    id_pelicula INTEGER,
    PRIMARY KEY (id_actor, id_pelicula),
    FOREIGN KEY (id_actor) REFERENCES actor(id_actor),
    FOREIGN KEY (id_pelicula) REFERENCES pelicula(id_pelicula)
);
```

### Tabla  "Premio_mejor_director"

```python
%%sql
CREATE TABLE premio_mejor_director (
    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    anio INTEGER,
    id_director INTEGER,
    FOREIGN KEY (id_director) REFERENCES director(id_director)
);
```

### Tabla  "Premio_mejor_actor"

```python
%%sql
CREATE TABLE premio_mejor_actor (
    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    anio INTEGER,
    id_actor INTEGER,
    FOREIGN KEY (id_actor) REFERENCES actor(id_actor)
);
```

### Tabla  "Premio_mejor_pelicula"

```python
%%sql
CREATE TABLE premio_mejor_pelicula ( 
    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    anio INTEGER,
    id_pelicula INTEGER,
    FOREIGN KEY (id_pelicula) REFERENCES pelicula(id_pelicula)
);
```

### Consulta 5
**Mostrar peliculas que tengan 2 o mas actores/actrices ordenadas de mayor a menor segun la cantidad de actores/actrices.**


**Ejemplo:**
|Nombre     |Cantidad de actores|
|-----------|-------------------|
|Oppenheimer|7                  |
|Poor Things|4                  |
|...        |...                |


**Solución:** _Se selecciona el nombre de las peliculas y se inicializa un contador que referenciara a la cantidad de actores, generando una relación entre la tabla actor participa en pelicula y pelicula, tomando comó punto en común el ID de la pelicula, luego la tabla actor participa en pelicula se relaciona con la tabla actor tomando comó punto en común el ID del actor, el resultado se agrupa por el nombre de pelicula siempre y cuando el contador sea mayor igual a 2, y finalmente se ordena por cantidad de actores en orden descendiente._

```python
%%sql
/* Reemplace estas queries por las que resuelven las consultas pedidas. */
SELECT p.nombre, COUNT(*) as cantidad_actores
FROM actor_participa_en_pelicula app
JOIN pelicula p ON app.id_pelicula = p.id_pelicula
JOIN actor a ON app.id_actor = a.id_actor
GROUP BY p.nombre
HAVING COUNT(*) >= 2
ORDER BY cantidad_actores DESC;
```

### Consulta 6
**Nombre de la película con la mayor brecha de edad entre el actor más joven y el actor más viejo que participen en ella, y las edades de dichos actores.** (Si hay empate, mostrar solamente uno).


**Ejemplo:**
|Nombre|Edad del actor más joven|Edad del actor más viejo|
|------|------------------------|------------------------|
|Barbie|16                      |92                      |


**Solución:** **_Se seleccionan los nombres de las peliculas y se calcula la edad minima, y la edad maxima de los actores que participan en la pelicula para obtener la diferencia de edad, esto mediante la extracción de datos de la tabla actor_participa_en_pelicula, generando uan relación central entre está tabla y la tabla pelicula, tomando comó punto en común el ID de la pelicula, luego se relaciona la tabla actor_participa_en_pelicula con la tabla actor, tomando comó punto en común el ID del actor, el resultado se agrupa por nombre y se ordena por la brecha de edad en orden decendiente, luego se selecciona el primer dato de la tabla resultante para obtener la mayor brecha de edad._**

```python
%%sql
/* Reemplace estas queries por las que resuelven las consultas pedidas. */
SELECT p.nombre AS nombre_pelicula,
    MAX(a.edad) - MIN(a.edad) AS brecha_edad,
    MIN(a.edad) AS edad_mas_joven,
    MAX(a.edad) AS edad_mas_viejo
FROM actor_participa_en_pelicula app
INNER JOIN pelicula p ON app.id_pelicula = p.id_pelicula
INNER JOIN actor a ON app.id_actor = a.id_actor
GROUP BY p.nombre
ORDER BY brecha_edad DESC
LIMIT 1;
```

### Consulta 7
**Año de cada premio ganado por la película de mayor duración.** (Si hay empate, elegir solamente uno).


**Ejemplo:**
|Año |

|2024|

**Solución:** **_Se selccionan los años de los premios a la mejor pelicula de la tabla premio_mejor_pelicula relacionada ccentralmente con la tabla pelicula, tomando como punto en común el ID de la pelicula, donde la duración de la pelicula es la mayor, el resultado se ordena por los años de forma descendente, distinct generara que el resultado entregue peliculas distintas, por lo cual entregara solo una pelicula que tenga la máxima duración._**

```python
%%sql
/* Reemplace estas queries por las que resuelven las consultas pedidas. */
SELECT DISTINCT pmp.anio
FROM premio_mejor_pelicula pmp
INNER JOIN pelicula p ON pmp.id_pelicula = p.id_pelicula
WHERE p.duracion = (
    SELECT MAX(duracion)
    FROM pelicula
)
ORDER BY pmp.anio DESC
LIMIT 1;
```

**Nombre de todos los actores que han trabajado con más de un director, la cantidad de directores con las que han trabajado y el nombre de una de las películas.**


**Ejemplo:**
|Nombre del actor|Cantidad de directores|Nombre de una película|

**Solución:** **_Esta consulta selecciona el nombre del actor, cuenta la cantidad de directores distintos con los que ha trabajado y selecciona el nombre de la película más antigua en la que ha participado, se realiza una unión entre las tablas actor_participa_en_pelicula, pelicula y actor, agrupando por el ID del actor y filtrando aquellos con más de un director. El resultado se ordena en orden descendente según la cantidad de directores._**

```python
%%sql
/* Reemplace estas queries por las que resuelven las consultas pedidas. */
SELECT a.nombre, COUNT(DISTINCT p.id_director) as cantidad_directores, MIN(p.nombre) as nombre_pelicula
FROM actor_participa_en_pelicula app
INNER JOIN pelicula p ON app.id_pelicula = p.id_pelicula
INNER JOIN actor a ON app.id_actor = a.id_actor
GROUP BY a.id_actor
HAVING COUNT(DISTINCT p.id_director) > 1
ORDER BY cantidad_directores DESC;
```

```python
%%sql
/* Reemplace estas queries por las que resuelven las consultas pedidas. */
SELECT a.nombre
FROM actor a
JOIN actor_participa_en_pelicula app ON a.id_actor = app.id_actor
JOIN pelicula p ON app.id_pelicula = p.id_pelicula
WHERE p.id_director = 1
GROUP BY a.id_actor, a.nombre
HAVING COUNT(*) = (
    SELECT COUNT(*)
    FROM actor_participa_en_pelicula app2
    JOIN pelicula p2 ON app2.id_pelicula = p2.id_pelicula
    WHERE app2.id_actor = a.id_actor
    AND p2.id_director = 1
);
```
